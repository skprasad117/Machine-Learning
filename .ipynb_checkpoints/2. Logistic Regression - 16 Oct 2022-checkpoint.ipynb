# some columns cant have 0 like glucose, insuline, skin thickness , BMI and blood pressure. either remove such dat or simply replace it with there mean value

```python
# replcing 0 with the mean of the column
for column in ['BMI','BloodPressure','Glucose','Insulin','SkinThickness']:
  if data[data[column]==0][column].count()>0:
    column_mean=data[column].mean()
    data[column] = data[column].replace(0,column_mean)
    print(column," column 0 values is corrected by replacing it with its mean :",column_mean)
```

```python
#all are numeric column. checking distribution of the data
plt.figure(figsize=(20,25), facecolor = 'white')
plotnumber = 1

for column in data:
  if plotnumber <=9:
    ax = plt.subplot(3,3,plotnumber)
    sns.distplot(data[column])
    plt.xlabel(column,fontsize=20)
    #plt.ylabel('',fontsize=20)
  plotnumber += 1
plt.show()
```

```python
q = data['Pregnancies'].quantile(0.98)
# we are removing the top 2% data from the pregnancies column
data_cleaned = data[data['Pregnancies']<q]
```

```python
temp=pd.DataFrame({"before_outliner":data['Pregnancies'],"after_outliner":data_cleaned['Pregnancies']})
```

```python
# visualizatio of pregancy data before and after handling outliner 
fig, ax = plt.subplots(figsize=(15,10))
sns.boxplot(data=temp,width = 0.5,ax=ax,fliersize=3)
```

```python
# similarlity handling other feature for outliner
# randomly removing few percent of data based on obervation
q = data_cleaned['BMI'].quantile(0.99)
data_cleaned = data_cleaned[data_cleaned['BMI']<q]

q = data_cleaned['SkinThickness'].quantile(0.99)
data_cleaned = data_cleaned[data_cleaned['SkinThickness']<q]

q = data_cleaned['Insulin'].quantile(0.95)
data_cleaned = data_cleaned[data_cleaned['Insulin']<q]

q = data_cleaned['DiabetesPedigreeFunction'].quantile(0.99)
data_cleaned = data_cleaned[data_cleaned['DiabetesPedigreeFunction']<q]

q = data_cleaned['Age'].quantile(0.99)
data_cleaned = data_cleaned[data_cleaned['Age']<q]
```

```python
print(data.shape)
print(data_cleaned.shape)
```

```python
#checking outlier
fig, ax = plt.subplots(figsize=(15,10))
sns.boxplot(data=data_cleaned,width = 0.5,ax=ax,fliersize=3)
```

```python
X=data_cleaned.drop(columns=['Outcome'])
X.head(2)
```

```python
y=data_cleaned['Outcome']
y.head()
```

```python
# 
plt.figure(figsize=(20,25),facecolor='white')
plotnumber = 1
for column in X:
  if plotnumber<=9:
    ax=plt.subplot(3,3,plotnumber)
    sns.stripplot(y,X[column])
  plotnumber +=1
plt.tight_layout()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
# multicollinerity
#Multicollinearity is the occurrence of high intercorrelations among 
#two or more independent variables in a multiple regression model. Multicollinearity can lead to skewed or misleading results
vif = pd.DataFrame()
vif["vif"] = [variance_inflation_factor(X_scaled,i)for i in range(X_scaled.shape[1])]
vif['Features'] = X.columns

vif
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y,test_size=0.25,random_state=355)
```

```python
log_reg = LogisticRegression()

log_reg.fit(X_train,y_train)
```

```python
## model saving

import pickle

# writing diffrent model files to file
with open('modelforprediction.sav','wb') as f:
  pickle.dump(log_reg,f)

with open('standardscaler.sav','wb') as f:
  pickle.dump(scaler,f)
```

```python
# prediction

y_pred = log_reg.predict(X_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

# confision matrix

```python
conf_mat = confusion_matrix(y_test,y_pred)
conf_mat
```

```python
true_positive = conf_mat[0][0]
false_positive = conf_mat[0][1]
false_negative  = conf_mat[1][0]
true_negative  = conf_mat[1][1]
```

```python
# accuracy formula
acc = (true_positive + true_negative)/(true_positive + true_negative+false_positive + false_negative)
acc
```

```python
precision = (true_positive)/(true_positive +  false_positive)
```

```python
precision
```

```python
recall = true_positive/(true_positive + false_negative)
recall
```

```python
f1_score = 2*(recall*precision)/(recall+precision)
```

```python
f1_score
```

```python
# roc auc
auc = roc_auc_score(y_test,y_pred)
auc
```

```python
fpr, tpr, thresholds = roc_curve(y_test,y_pred)
```

```python
plt.plot(fpr, tpr, color = 'orange', label='ROC')
plt.plot([0,1],[0,1],color='darkblue', linestyle='--', label='ROC Curve(area= %0.2f)'% auc)
plt.xlabel('True Positie Rate')
plt.ylabel('false positive rate')
plt.title('reciver operating characterstic (ROC) Curve')
plt.legend()
plt.show()
```
